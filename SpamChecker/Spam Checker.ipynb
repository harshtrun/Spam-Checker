```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('spam.csv',encoding='latin-1')
```

```python
df.head()
```

```python
#with open opens a file. 'r/w' is to set permission for reading/write. strip() removes returns from the file text.
# <_io.TextIOWrapper name='stopwords.txt' mode='r' encoding='cp1252'> is actually the object printed here.
```

```python
stoplist=[]
with open('stopwords.txt','r') as x:
    #prints the object
    print(x)
    for i in x:
        stoplist.append(i.strip())
```

```python
dataList=[]
for mail in df.v2:
    sentlist=[]
    for sent in mail.strip().split('.'):
        for word in sent.split():
            if word not in stoplist:
                sentlist.append(word)
#wordlist.append(word)
#if len(wordlist)!=0:
    dataList.append(sentlist)
```

```python
 dataList
```

```python
#step to remove single freq wrds from the datalist
#defaultdictionary creates a dictionary with every element and it freq.
from collections import defaultdict
freqwords=defaultdict(int)
for mail in dataList:
    for word in mail:
        freqwords[word]+=1
```

```python
freqwords['point,']
```

```python
#Final Filtered Data
filtData=[]
for mail in dataList:
    filtMail=[]
    for word in mail:
        if freqwords[word]>1:
            filtMail.append(word)
    filtData.append(filtMail)
```

```python
filtData
```

```python
import gensim
```

```python
dictBow=gensim.corpora.Dictionary(filtData)
bowData=[dictBow.doc2bow(mail) for mail in filtData]
n_unique_tokens=len(dictBow)
dense_bow=gensim.matutils.corpus2dense(bowData,num_terms=n_unique_tokens).transpose()
dense_bow
```

```python
dimension_vectors=n_unique_tokens
dimension_vectors
n_unique_tokens
```

```python
#Breaking the data into testing and training sets
l1=len(dense_bow)
l1
```

```python
p1=0.9*l1
p1=int(p1)
p1
```

```python
set_train=dense_bow[0:p1]
set_test=dense_bow[p1:l1]
len(set_train)
```
